```
@app.route('/search',methods=['POST'])
def search():
  file = request.files['query_img']
  img = Image.open(file.stream)  # PIL image 
  uploaded_img_path = 'Datasets/webapp/' + file.filename
  img.save(uploaded_img_path)

  query_output = [{
                    "geo_coordinates": {"lat":1.23,"long":4.56},
                    "score": 5.67,
                    "image_url": "/file?path="+uploaded_img_path
                  },
                  {
                    "geo_coordinates": {"lat":1.23,"long":4.56},
                    "score": 5.67,
                    "image_url": "/file?path="+uploaded_img_path
                  },
                  {
                    "geo_coordinates": {"lat":1.23,"long":4.56},
                    "score": 5.67,
                    "image_url": "/file?path="+uploaded_img_path
                  },
                  {
                    "geo_coordinates": {"lat":1.23,"long":4.56},
                    "score": 5.67,
                    "image_url": "/file?path="+uploaded_img_path
                  }
                  ]
  json_object = json.dumps(query_output)
  return json_object
```

```
@app.route('/file',methods=['GET'])
def file():
  path = request.args.get('path')
  return send_file(ROOT + '/' + path, mimetype='image/jpeg')
```
